```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.rename(columns={'movie_id': 'id'}, inplace = True)
credits.columns
```

```python
df = pd.merge(movies, credits, on = 'id')
df.drop(['homepage', 'production_countries', 'status', 'status', 'title_x', 'title_y'], 1, inplace = True)
df.columns
```

```python
df.head()
```

# Content based Recommended System

    In this we are using movie summery for perdiction

```python
df.head(1)['overview']
```

 Creating document matrics from each of the sentences

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(min_df=3,strip_accents='unicode', 
                     analyzer = 'word',
                     token_pattern='\w{1,}',
                     ngram_range=(1, 3),
                    stop_words='english')

df['overview'] = df['overview'].fillna('') 
```

```python
#  It create a sparse matrix (which has lot of 0z)
tf_matrix = tf.fit_transform(df['overview'])
tf_matrix.shape
```

# comparing each movies with one another(using Sigmoid)

           1 2 3
         1
         2 
         3 


```python
from sklearn.metrics.pairwise import sigmoid_kernel

sf_k = sigmoid_kernel(tf_matrix, tf_matrix)
```

```python
sf_k[0]
```
